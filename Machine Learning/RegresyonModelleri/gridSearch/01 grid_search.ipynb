# Grid Search

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("diyabet.xlsx")
```

```python
df.head()
x = df.drop("S6",axis=1)
y = df["S6"]

```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
   x, y, test_size=0.3, random_state=9)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## Grid Search

1. Tahmin edilecek bir algoritma kullan
2. Parametre belirle
3. Verileri grid search e oturt
4. Cv unutma
5. Puan Fonksiyonu Belirle

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet = ElasticNet()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    "alpha" : [.01 , .1 , 1 , 5 ,10 , 50 , 100 , 1000] ,  
    "l1_ratio":[.001 , 0.01 , 0.1 , .5  , .7 , .9 , .95 , .99]
    
    }
```

```python
grid_model = GridSearchCV(
    estimator=elasticnet,
    param_grid=params,
    scoring="neg_mean_squared_error",
    cv=5,
    verbose=2
    )
```

```python
grid_model.fit(X_train , y_train)
```

```python
grid_model.best_estimator_ # En iyi değer
```

```python
grid_result = pd.DataFrame(grid_model.cv_results_)
```

```python
grid_result
```

```python
y_pred = grid_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
np.sqrt(mean_squared_error(y_test , y_pred))
```
