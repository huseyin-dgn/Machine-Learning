# Train Test Split

```python
x = df[["age"]]
y = df[["visits"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
help(train_test_split)
```

 Train_test_split fonksiyonu :  Veriyi rastgele eğitim ve test kümelerine böler.

x: Bağımsız değişkenler (modelin öğrenmesini sağlayan özellikler).

y: Bağımlı değişken (modelin tahmin etmeye çalıştığı değerler).

test_size=0.3: Verinin %30’unu test, %70’ini eğitim için ayırır.

random_state=9: Rastgele veri bölme işleminin her çalıştırmada aynı sonucu vermesini sağlar.

X_train, y_train: Modelin eğitileceği veri kümesi.

X_test, y_test: Modelin test edileceği veri kümesi.

```python
X_train, X_test, y_train, y_test = train_test_split(
        x, y, test_size=0.3 , random_state=99)
```

# Scikit Learn


```python
from sklearn.linear_model import LinearRegression
```

```python
regr= LinearRegression()
```

regr.fit(X_train, y_train) → Modeli eğitim verisiyle eğitir.

X_train → Bağımsız değişkenler (özellikler/features).

y_train → Bağımlı değişkenler (etiketler/labels).

Model, X_train ve y_train verilerini öğrenerek bir 
fonksiyon oluşturur ve gelecekteki tahminleri buna göre yapar.

```python
regr.fit(X_train , y_train) # Eğitilmek istenilen değerler
```

.corr() fonksiyonu, iki değişken arasındaki Pearson korelasyon katsayısını hesaplar.


Sonuç -1 ile 1 arasında bir değer olur:
Pozitif Korelasyon (+1’e yakın) → Bir değişken artarken diğeri de artıyorsa (örneğin, yaş arttıkça ziyaret sayısı da artıyorsa).

Negatif Korelasyon (-1’e yakın) → Bir değişken artarken diğeri azalıyorsa (örneğin, yaş arttıkça ziyaret sayısı azalıyorsa).

0’a Yakın Korelasyon → İki değişken arasında anlamlı bir ilişki yoktur.

```python
df["age"].corr(df["visits"])
```

# Performans

```python
test_pred = regr.predict(X_test)

# regr.predict(X_test), daha önce eğitilmiş modelin (regr) X_test veri kümesi için tahmin yapmasını sağlar.
```

```python
test_pred
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
```

```python
print("Theta :     ", (regr.coef_[0]))
print("Intercept : ", (regr.intercept_))
print("R^2 :       ", round(r2_score(y_test,test_pred),2))
print("MAE :       ", round(mean_absolute_error(y_test,test_pred),2))
print("RMSE :      ", round(np.sqrt(mean_squared_error(y_test,test_pred)),2))
```

print("Theta : ", (regr.coef_[0])) :  Modelin bağımsız değişkenin katsayısı (özelliklerin etkisi).


print("Intercept : ", (regr.intercept_)) : Modelin kesişim (intercept) değeri, yani y-ekseniyle kesişim noktası.


print("R^2 : ", round(r2_score(y_test, test_pred), 2)) :  Modelin doğruluğunu gösteren R² değeri (verinin model tarafından ne kadar açıklandığı).


print("MAE : ", round(mean_absolute_error(y_test, test_pred), 2)) : MAE (Mean Absolute Error): Modelin tahmin hatasının ortalama mutlak değeri.


print("RMSE : ", round(np.sqrt(mean_squared_error(y_test, test_pred)), 2)) : RMSE (Root Mean Squared Error): Modelin tahmin hatalarının karekökü.
