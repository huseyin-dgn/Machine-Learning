```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine

from sklearn.metrics import r2_score , mean_absolute_error , mean_squared_error
```

```python
veri = load_wine()

df= pd.DataFrame(data = veri.data , columns=veri.feature_names)

df["proline"]= veri.target
```

```python
df.describe()
```

```python
x= df.drop("proline" , axis=1)
y= df["proline"]
```

# Polinom Dönüşümü

```python
from sklearn.preprocessing import PolynomialFeatures

poly_conv = PolynomialFeatures(degree=2 , include_bias=False)

poly_feat= poly_conv.fit_transform(x)
```

# TRAİN TEST SPLİT

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , y_train , y_test = train_test_split(x,y,test_size=0.2 , random_state=9)
```

# STANDART SCALER

ortalama =  0 
standart sapma = 1 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

# RİGDE REGRESYON
bias = veri kaybını önlemek

varyans = verideki dağınıklılığı önlemek

yüksek varyans = overfitting

yüksek bias = underfitting



```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha =1)
```

```python
ridge_model.fit(X_train , y_train)
```

```python
ridge_pred = ridge_model.predict(X_test)
```

```python
print("R^2 : " , round(r2_score(y_test , ridge_pred),2))
print("MAE : " , round(mean_absolute_error(y_test , ridge_pred),2))
print("RMSE : " , round(mean_squared_error(y_test , ridge_pred),2))
```

# LASSO

```python
# Gerekli kütüphaneyi içe aktarıyoruz
from sklearn.linear_model import LassoCV
# LassoCV modelini oluşturuyoruz
lasso_cv = LassoCV(
    eps=0.1,        # Minimum ve maksimum alpha değerleri arasındaki oran (eps * alpha_max = min_alpha)
    n_alphas=100,   # Logaritmik olarak seçilecek alpha (ceza parametresi) sayısı
    cv=5            # 5 katlı çapraz doğrulama (veriyi 5 parçaya bölerek en iyi alpha'yı seçer)
)
```

```python
lasso_cv.fit(X_train , y_train)
```

```python
lasso_cv.alpha_
```

```python
lasso_cv.coef_

# bazıları 0 olacak.
# lasso diyor ki bunlar benim işime yaramaz 0 yaptım
```

```python
lasso_pred = lasso_cv.predict(X_test)
```

```python
print("R^2 : " , round(r2_score(y_test , lasso_pred),2))
print("MAE : " , round(mean_absolute_error(y_test , lasso_pred),2))
print("RMSE : " , round(mean_squared_error(y_test , lasso_pred),2))
```
